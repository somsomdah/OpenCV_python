```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

# cv2.findChessbordCorners, cv2.drawChessbordCorners

```python
src=cv2.imread('./img/chessboard.jpg')
gray=cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
pattern_size=(6,3)
```

```python
found,corners=cv2.findChessboardCorners(src,pattern_size)
dst=src.copy()
dst=cv2.drawChessboardCorners(dst,pattern_size,corners,found)
```

```python
plt.figure(figsize=(15,30))

plt.subplot(1,2,1)
plt.imshow(gray,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(dst)

plt.show()
```


  
  
# cv2.findCirclesGrid

```python
src=cv2.imread('./img/circle_grid.jpg')
gray=cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
pattern_size=(6,3)
```

```python
found,centers=cv2.findCirclesGrid(src,pattern_size)
dst=src.copy() 
dst=cv2.drawChessboardCorners(dst,pattern_size,centers,found) # dst도 변형시킴
```

```python
plt.figure(figsize=(15,30))

plt.subplot(1,2,1)
plt.imshow(src)

plt.subplot(1,2,2)
plt.imshow(dst)

plt.show()
```
