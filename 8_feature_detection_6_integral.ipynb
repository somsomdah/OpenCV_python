```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.arange(1, 17).reshape(4, 4).astype(np.uint8)
sumA, sqsumA, tiltedA = cv2.integral3(A)

A=cv2.normalize(A, None, 0, 255, cv2.NORM_MINMAX)
sumA=cv2.normalize(A, None, 0, 255, cv2.NORM_MINMAX)
sqsumA=cv2.normalize(sqsumA, None, 0, 255, cv2.NORM_MINMAX)
tiltedA=cv2.normalize(tiltedA, None, 0, 255, cv2.NORM_MINMAX)
```

```python
plt.figure(figsize=(20,80))

plt.subplot(1,4,1)
plt.imshow(A,cmap='gray')

plt.subplot(1,4,2)
plt.imshow(sumA,cmap='gray')

plt.subplot(1,4,3)
plt.imshow(sqsumA,cmap='gray')

plt.subplot(1,4,4)
plt.imshow(tiltedA,cmap='gray')

plt.show()
```

```python
gray = cv2.imread('./img/Lenna.png', cv2.IMREAD_GRAYSCALE)
gray_sum = cv2.integral(gray)
dst = cv2.normalize(gray_sum, None, 0, 255,cv2.NORM_MINMAX, dtype=cv2.CV_8U)
```

```python
plt.figure(figsize=(20,80))

plt.subplot(1,3,1)
plt.imshow(gray,cmap='gray')

plt.subplot(1,3,2)
plt.imshow(gray_sum,cmap='gray')

plt.subplot(1,3,3)
plt.imshow(dst,cmap='gray')

plt.show()
```
