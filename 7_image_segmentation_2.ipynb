```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Image Segmentation by Color : cv2.InRange

```python
hand=cv2.imread('hand.jpg')
lenna=cv2.imread('lenna.png')
soda=cv2.imread('soda_can.jpg')
```

```python
hand=cv2.cvtColor(hand,cv2.COLOR_BGR2HSV)
lenna=cv2.cvtColor(lenna,cv2.COLOR_BGR2HSV)
soda=cv2.cvtColor(soda,cv2.COLOR_BGR2HSV)

# image histogram으로 확인해야함...
lowerb1=(0,40,0)
upperb1=(20,180,255)

dst1=cv2.inRange(hand,lowerb1,upperb1)
dst2=cv2.inRange(lenna,lowerb1,upperb1)
dst3=cv2.inRange(soda,lowerb1,upperb1)
```

```python
plt.figure(figsize=(20,60))

plt.subplot(1,3,1)
plt.imshow(dst1,cmap='gray')

plt.subplot(1,3,2)
plt.imshow(dst2,cmap='gray')

plt.subplot(1,3,3)
plt.imshow(dst3,cmap='gray')

plt.show()
```

# Finding Contours : cv2.findContours

- 다각형만 검출 가능

```python
hand=cv2.imread('hand.jpg',cv2.IMREAD_GRAYSCALE)
polygons=cv2.imread('polygons.png',cv2.IMREAD_GRAYSCALE)

contours1,hirearchy1=cv2.findContours(hand,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
contours4,hirearchy4=cv2.findContours(polygons,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)

hand=cv2.drawContours(hand,contours1,-1,(0,255,0),3)
polygons=cv2.drawContours(polygons,contours4,-1,(0,255,0),3)
```

```python
plt.figure(figsize=(40,80))

plt.subplot(1,2,1)
plt.imshow(hand)

plt.subplot(1,2,2)
plt.imshow(polygons)

plt.show()
```
