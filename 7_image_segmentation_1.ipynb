```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Canny edge detector

```python
src=cv2.imread('Lenna.png',cv2.IMREAD_COLOR)
src=cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
edges1=cv2.Canny(src,50,100)
edges2=cv2.Canny(src,50,200)
```

```python
plt.figure(figsize=(40,20))

plt.subplot(1,2,1)
plt.imshow(edges1,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(edges2,cmap='gray')

plt.show()
```

# Hugh Transform

```python
sudoku=cv2.imread('sudoku.png',cv2.IMREAD_COLOR)
sudoku_gray=cv2.cvtColor(sudoku,cv2.COLOR_BGR2GRAY)
sudoku_edges=cv2.Canny(sudoku_gray,50,100)
sudoku_lines=cv2.HoughLines(sudoku_edges,rho=1,theta=np.pi/180.0,threshold=100)


for line in sudoku_lines:

    rho,theta=line[0]
    c=np.cos(theta)
    s=np.sin(theta)
    x0=c*rho
    y0=s*rho
    x1 = int(x0 + 1000*(-s))
    y1 = int(y0 + 1000*(c))
    x2 = int(x0 - 1000*(-s))
    y2 = int(y0 - 1000*(c))
 
    cv2.line(sudoku,(x1,y1),(x2,y2),(0,0,255),1)
```

```python
plt.figure(figsize=(40,20))

plt.subplot(1,2,1)
plt.imshow(sudoku)

plt.subplot(1,2,2)
plt.imshow(sudoku_edges,cmap='gray')

plt.show()
```

```python
sudoku2=cv2.imread('sudoku.png',cv2.IMREAD_COLOR)
lines=cv2.HoughLinesP(sudoku_edges,rho=1,theta=np.pi/180,threshold=100)

for line in lines:
    x1,y1,x2,y2=line[0]
    cv2.line(sudoku2,(x1,y1),(x2,y2),(0,0,255),2)
```

```python
plt.figure(figsize=(40,20))

plt.subplot(1,2,1)
plt.imshow(sudoku2)

plt.subplot(1,2,2)
plt.imshow(sudoku_edges,cmap='gray')

plt.show()
```

```python
soda=cv2.imread('soda_can.jpg')
gray=cv2.cvtColor(soda,cv2.COLOR_BGR2GRAY)

edges=cv2.Canny(gray,50,100)
circles=cv2.HoughCircles(gray,method=cv2.HOUGH_GRADIENT,dp=1,minDist=100)

for circle in circles[0,:]:
    cx,cy,r=circle
    cv2.circle(soda,(cx,cy),r,(0,0,255),2)
```

```python
plt.figure(figsize=(40,20))

plt.subplot(1,2,1)
plt.imshow(soda)

plt.subplot(1,2,2)
plt.imshow(edges,cmap='gray')

plt.show()
```
